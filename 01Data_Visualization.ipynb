```python
import pandas
```

```python
import matplotlib.pyplot as plt
```

```python
dict={
    "name":["Gaurav","Mithilesh",'Harsh','Virat','Aditya'],
    'english':[85,63,48,95,66],
    'maths':[88,96,15,23,63],
    'science':[12,67,78,69,45],
    'marathi':[99,20,52,25,96]
}
```

```python
df=pandas.DataFrame(dict)
print(df)
```

Data Visualization Line Chart

```python
plt.plot(df['name'],df['english'])
```

```python
plt.plot(df['name'],df['maths'])
plt.title('marks')
plt.xlabel('name')
plt.ylabel('maths')
plt.grid(True)
plt.show()
```

```python
plt.plot(df['name'],df['marathi'])
plt.title("marks")
plt.xlabel("name")
plt.ylabel("marathi")
plt.grid(True)
plt.show()
```

```python
plt.plot(df['name'],df['maths'])
plt.title("Marks")
plt.xlabel("Name of Students")
plt.ylabel("Marks of Maths")
plt.grid(True)
plt.show()
```

**Data Visualization using Barchart**

```python
plt.bar(df['name'],df['maths'],color='red')
plt.title("Bar chart of marks")
plt.xlabel("Name of Students")
plt.ylabel("Marks of Maths")
plt.show()
```

```python
plt.bar(df['name'],df['maths'],color='blue')
plt.title("Bar chart of marks")
plt.xlabel("Name of Students")
plt.ylabel("Marks of Maths")
plt.show()
```

**Data Visualization using Scatter chart**

```python
plt.scatter(df['name'],df['english'])
plt.plot(df['name'],df['english'],color='black')
plt.title("Scatter chart of marks")
plt.xlabel("Name of Students")
plt.ylabel("Marks of English")
plt.show()
```

```python
plt.scatter(df['name'],df['marathi'],color='green')
plt.plot(df['name'],df['marathi'],color='blue')
plt.title("Scatter chart of marks")
plt.xlabel("Name of Students")
plt.ylabel("Marks of Marathi")
plt.show()
```

```python
plt.scatter(df['name'],df['science'],color='pink')
plt.plot(df['name'],df['science'],color='orange')
plt.title("Scatter chart of marks")
plt.xlabel("Name of Students")
plt.ylabel("Marks of Marathi")
plt.show()
```

**Data Visualization using Histogram **

```python
plt.hist(df['english'],color='pink',edgecolor='black')
plt.title("Histogram of marks")
plt.xlabel("Marks of English")
plt.ylabel("Frequency")
plt.show()
```

**Data Visualization using piechart**

```python

data={
    'year':[2019,2020,2021,2022,2023,2024],
    'sales':[345,296,189,380,335,396]
}
```

```python
dt=pandas.DataFrame(data)
print(dt)
```

```python
sizes=dt['sales']
label=dt['year']
```

```python
plt.pie(sizes,labels=label,autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(sizes,labels=label,autopct="%1.1f%%")
plt.show()
```
