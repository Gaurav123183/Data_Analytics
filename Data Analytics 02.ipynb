```python
url="https://raw.githubusercontent.com/sohamglobal/datasets/refs/heads/main/Flipkart_Mobiles.csv"
df=pandas.read_csv(url)
print(df)
```

```python
count=df['Brand'].value_counts()
print(count)
```

```python
df['Rating'].value_counts(ascending=False).head(10)
```

```python
df[["Brand","Rating"]].sort_values(by="Rating",ascending=False).head(10)
```

```python
ans=df[(df['Color']=="Moonlight Black") & (df['Selling Price']<=20000)]
print(ans)
```

```python
percentage=count/(count.sum())*100
```

```python
data=pandas.DataFrame({"Count":count,"Percentage":percentage})
print(data)
```
