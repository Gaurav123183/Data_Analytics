```python
import pandas
```

```python
import seaborn
```

```python
import matplotlib.pyplot as plt
```

```python
dict={
    'player':['Salah','Trent','Darwin','Gakpo','Jota','Luis'],
    'goals':[29,3,5,10,6,13]
}
```

```python
df=pandas.DataFrame(dict)
print(df)
```

**Line Plot Using Seaborn**

```python
seaborn.lineplot(x='player',y='goals',data=df,linestyle="--",color="red",linewidth=2)
plt.title('Score')
plt.show()
```

```python
seaborn.barplot(x='player',y='goals',data=df,color='indigo',edgecolor='black')
plt.title("Score")
# plt.grid()
plt.show()
```

```python
seaborn.barplot(x='goals',y='player',data=df,edgecolor='black')
plt.title("Score")
plt.show()
```

```python
seaborn.barplot(x='goals',y='player',data=df,edgecolor='black')
plt.title("Score")
plt.grid()
plt.show()
```

```python
seaborn.scatterplot(x='player',y='goals',data=df,color='indigo',edgecolor='black')
plt.plot(df['player'],df['goals'],linestyle=':',color='black')
plt.show()
```

```python
seaborn.scatterplot(x='player',y='goals',data=df,color='skyblue',edgecolor='black')
plt.plot(df['player'],df['goals'],linestyle=':',color='black')
plt.grid()
plt.show()
```

```python
seaborn.histplot(x='goals',data=df,color='orange',edgecolor='black')
plt.show()
```
