```python
url="CancerData.csv"
```

```python
df=pandas.read_csv(url)
print(df)
```

```python
df.rename(columns={"Cancer_Type":"cancer"})
df.rename(columns={"Treatment_Type":"Treatment"})
df.rename(columns={"Patient_ID":"ID"})
```
